```python
import pandas as pd
```

```python
# Create vine_table dataframe

vine_table_df = pd.read_csv("vine_table.csv")

vine_table_df.head()
```

```python
# Create filtered data frame where total_votes >= 20

large_total_votes_df = vine_table_df[vine_table_df["total_votes"] >= 20]

large_total_votes_df.head()
```

```python
# Create filtered data frame where helpful_votes divided by total_votes >= 0.5

other_total_votes_df = large_total_votes_df[large_total_votes_df["helpful_votes"]/large_total_votes_df["total_votes"] >= 0.5]

other_total_votes_df.head()
```

```python
# Create filtered data frame for reviews on vine program

vine_reviews_df = other_total_votes_df[other_total_votes_df["vine"] == 'Y']

vine_reviews_df
```

```python
# Create filtered data frame for reviews no on vine program

non_vine_reviews_df = other_total_votes_df[other_total_votes_df["vine"] == 'N']

non_vine_reviews_df
```

```python
# Calculate total, total 5-star reviews, and 5-star percentage

vine_count = vine_reviews_df["review_id"].count()

vine_five_df = vine_reviews_df[vine_reviews_df["star_rating"] == 5]

vine_five_count = vine_five_df["review_id"].count()

five_percent = (vine_five_count/vine_count) * 100

# Create Dataframe summarizing findings

vine_five_df = pd.DataFrame({"total_votes": vine_count, "five_star_count": vine_five_count, "five_star_percent": five_percent}, index = [0])

vine_five_df
```

```python
# Calculate total, total 5-star reviews, and 5-star percentage

non_vine_count = non_vine_reviews_df["review_id"].count()

non_vine_five_df = non_vine_reviews_df[non_vine_reviews_df["star_rating"] == 5]

non_vine_five_count = non_vine_five_df["review_id"].count()

non_five_percent = (non_vine_five_count/non_vine_count) * 100

# Create Data Frame summarizing findings

non_five_df = pd.DataFrame({"total_votes": non_vine_count, "five_star_count": non_vine_five_count, "five_star_percent": non_five_percent}, index = [0])

non_five_df
```
